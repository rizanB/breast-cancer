Label encoding the diagnosis column so it doesn't have string values - "benign" and "malignant"

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
dataset["diagnosis"] = le.fit_transform(dataset["diagnosis"])
```
